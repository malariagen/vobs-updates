The following sample sets in the Ag3.2 dataset have been released from publication embargo today:

```python
import malariagen_data
ag3 = malariagen_data.Ag3(show_progress=False)
ag3
```

```python
def study_link(value):
    return f'<a href="https://www.malariagen.net/partner_study/{value}">{value}</a>'

(
    ag3
    .sample_sets(release="3.2")
    .query("terms_of_use_expiry_date == '2024-07-22'")
    [["sample_set", "sample_count", "study_id", "release", "terms_of_use_expiry_date"]]
    .style
    .hide(axis="index")
    .format(formatter=study_link, subset="study_id")
)
```

These sample sets all originate from the [GAARD project](https://www.anophelesgenomics.org/#/home) and have been released from publication embargo following publication of results in [Lucas et al. (2023)](https://doi.org/10.1038/s41467-023-40693-0).

The sample sets include samples from Benin, Togo, Ghana and Cote d'Ivoire. Here is a summary of the number of samples by location, year and taxon:

```python
sample_sets = [
    "1237-VO-BJ-DJOGBENOU-VMF00050",
    "1237-VO-BJ-DJOGBENOU-VMF00067",
    "1244-VO-GH-YAWSON-VMF00051",
    "1245-VO-CI-CONSTANT-VMF00054",
    "1253-VO-TG-DJOGBENOU-VMF00052",
]
ag3.count_samples(
    sample_sets=sample_sets,
    index=["sample_set", "country", "admin1_iso", "admin1_name", "year"]
)
```
